```python
import Quartz as Qz
from Quartz import *

quartz_dir_list=dir(Qz)

import objc


def hex_str(v):
    return "{0:#x}".format(v)

def getMatchingQzEntries(match,case_sensitive=False):
    if case_sensitive is False:
        match=match.lower()
        return [i for i in quartz_dir_list if i.lower().find(match)>-1]
    else:
        return [i for i in quartz_dir_list if i.find(match)>-1]

def getQzMatchingConstants(match,case_sensitive=False):
    entries=getMatchingQzEntries(match,case_sensitive)
    return [(i,getattr(Qz,i)) for i in entries if callable(i) is False]

def printFormatedConstants(key_value_pairs, key_padding=36):
    print '\n'
    for k,v in key_value_pairs:
        uchar='N/A'
        if v < 0x10000:
            uchar=unichr(v)
        print u"{0} = {1} # {2}".format(k.ljust(key_padding),hex_str(v).ljust(10),uchar)
    print '\n'
```
